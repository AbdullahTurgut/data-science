```python
def alcohol_category(x):
    if x < 10:
        return "Low"
    elif 10 <= x < 12:
        return "Medium"
    else:
        return "High"

df["alcohol_category"] = df["alcohol"].apply(alcohol_category)
```

```python
df["quality_label"] = df["quality"].apply(lambda q: "Good" if q >= 7 else "Bad")
```

```python
df["acidity_level"] = df["pH"].apply(lambda x: "Low" if x < 3.2 else ("Medium" if x < 3.4 else "High"))
```

```python
df
```

```python
# Alkol kategorisine gÃ¶re kalite daÄŸÄ±lÄ±mÄ±
sns.countplot(x="alcohol_category", hue="quality_label", data=df, palette="Set1")
plt.title("Alkol Seviyesi vs Kalite")
plt.show()
```

```python
# Kaliteye gÃ¶re acidity_level
sns.countplot(x="acidity_level", hue="quality_label", data=df, palette="Set2")
plt.title("Asitlik Seviyesi vs Kalite")
plt.show()
```

```python
# Pie chart - Good vs Bad wines
df["quality_label"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=["lightcoral","lightgreen"])
plt.title("Ä°yi vs KÃ¶tÃ¼ ÅžaraplarÄ±n DaÄŸÄ±lÄ±mÄ±")
plt.ylabel("")  # y-axis yazÄ±sÄ±nÄ± kaldÄ±r
plt.show()
```
