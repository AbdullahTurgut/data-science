```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# jupiter notebook için görselleştirme yaparken önemli
```

```python
ageList = [10,20,30,30,30,40,60,70,75]
```

```python
weightList = [20,60,80,85,86,87,70,90,95]
```

```python
plt.plot(ageList,weightList,"g") # y -> yellow, g -> green, r -> red
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.savefig("test.png")
```

```python
plt.show()
```

```python
# numpy
```

```python
npAgeList = np.array(ageList)
npWeightList = np.array(weightList)
```

```python
plt.plot(npAgeList,npWeightList,"r") # y -> yellow, g -> green, r -> red
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
npArr = np.linspace(0,10,20)
```

```python
npArr
```

```python
npArr2 = npArr ** 3
```

```python
npArr2
```

```python
plt.plot(npArr,npArr2,"b*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(npArr,npArr2,"r*-")
plt.subplot(1,2,2)
plt.plot(npArr2,npArr,"b--")
plt.show()
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.1,0.1,0.4,0.4])
figureAxes.plot(npArr,npArr2,"g")
figureAxes.set_xlabel("X Axis")
figureAxes.set_ylabel("Y Axis")
figureAxes.set_title("Graph Title")
plt.show()
```

```python
newFig = plt.figure(dpi=100)
newAxes = newFig.add_axes([0.1,0.1,0.9,0.9])
newAxes.plot(npArr, npArr**2, label = "numpy array ** 2")
newAxes.plot(npArr, npArr**3, label = "numpy array ** 3")
newAxes.legend() # can locate -> loc=3 an example
plt.show()
```
