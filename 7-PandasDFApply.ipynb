```python
def SalaryCategory(salary):
    if salary <= 50000:
        return "Low"
    elif salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df['Salary'].apply(SalaryCategory)
```

```python
df["Salary_Category"] = df['Salary'].apply(SalaryCategory)
```

```python
# TODO: if experience > 10 -> performanceScore + 1
```

```python
def AddPerformanceScoreForExperience(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df['Updated_Performance_Score'] = df.apply(AddPerformanceScoreForExperience,axis=1) # axis 1 satÄ±r bazlÄ± iÅŸlem
```
