```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df[["sex","class","embark_town"]].isna().sum()
```

```python
df = df.dropna(subset=['embark_town'])
```

```python
df[["sex","class","embark_town"]].isna().sum()
```

```python
# one-hot encoding
```

```python
df['sex'].value_counts()
```

```python
df['embark_town'].value_counts()
```

```python
df['class'].value_counts()
```

```python
df_oneHot = pd.get_dummies(df, columns=["sex","embark_town"], drop_first=True)
```

```python
df_oneHot
```

```python
df_oneHot.columns
```

```python
# label encoder
```

```python
labelEncoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label["sex"] = labelEncoder.fit_transform(df_label["sex"])
```

```python
df_label
```

```python
# ordinal encoder
```

```python
df_ordinal = df.copy()
```

```python
ordinalEncoder = OrdinalEncoder(categories=[["Third","Second","First"]])
```

```python
df_ordinal["class"] = ordinalEncoder.fit_transform(df_ordinal[["class"]])
```

```python
df_ordinal
```

```python
fig , axes = plt.subplots(1,3,figsize=(15,5))
df["sex"].value_counts().plot(kind="bar",ax=axes[0], title="Original Categorical")
df_label["sex"].value_counts().plot(kind="bar",ax=axes[1], title="Label Encoded")
df_oneHot["sex_male"].value_counts().plot(kind="bar",ax=axes[2], title="OneHot Encoded")
```
